```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("https://raw.githubusercontent.com/rfordatascience/tidytuesday/master/data/2020/2020-02-18/food_consumption.csv")
```

```python
array = ['Kenya','Uganda','Tanzania','Rwanda','Ethiopia']
cat = data[data['country'].isin(array)]
```

```python
cat.groupby('food_category')['co2_emmission'].mean().reset_index().sort_values('co2_emmission')
```

```python
food_data = data.groupby('food_category')[['co2_emmission']].sum().sort_values(by="food_category",ascending=False).reset_index()
```

```python
food_data
```

```python
categories = food_data['food_category']
values =food_data['co2_emmission']
```

```python


plt.figure(figsize=(15,6))
plt.barh(categories,values,color='purple')
plt.title("food category by emmision",fontsize=15,fontweight='bold',loc='left',pad=15)
plt.xlabel("co2_emmission")
plt.grid(axis='x')
plt.show()
```

```python
sns.boxplot(data=data,y='food_category',x='co2_emmission')
plt.show()
```

```python
sns.boxplot(data=data,hue='food_category',x='co2_emmission')
plt.show()
```
